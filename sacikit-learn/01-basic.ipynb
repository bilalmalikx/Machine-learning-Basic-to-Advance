```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# simple linear regression
from sklearn.linear_model import LinearRegression

# load the data

df = sns.load_dataset('tips')
df.head()
```

```python
X = df[['total_bill']]
Y = df['tip']

model = LinearRegression()
model.fit(X, Y)
model.predict([[24.59]])
```
